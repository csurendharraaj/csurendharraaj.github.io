```python
import PyPDF2
from PyPDF2 import PdfReader

book=PdfReader("AU127-1.pdf")
np=len(book.pages)

txt=""
for i in range(np):
    txt+=book.pages[i].extract_text().lower()
    
print(txt)
```

```python
txt_l=txt.strip().splitlines()

para = ''
for txt in txt_l:
    para += txt
para
```
