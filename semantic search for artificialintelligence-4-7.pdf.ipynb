```python
import PyPDF2
from PyPDF2 import PdfReader

book=PdfReader("artificialintelligence-4-7.pdf")
np=len(book.pages)

txt=""
for i in range(np):
    txt+=book.pages[i].extract_text().lower()
    
print(txt)
```

```python
txt_l=txt.strip().splitlines()
print(txt_l)
para = ''
for txt in txt_l:
    para += txt
para
```

```python
txt_list=para.split('.')
txt_list
txt_l=[]
for i in txt_list:
    if(len(i.strip())!=0):
        txt_l.append(i.strip())
```

```python
embedding_model="text-embedding-ada-002"

e_list=[]
for i in txt_l:
    e_list.append(get_embedding(i,engine=embedding_model))

e_list
```

```python
print(len(txt_l),len(e_list))
```

```python
def search_query(query):
    query_e=get_embedding(query,engine=embedding_model)
    
    similarity=[]
    for i in e_list:
        similarity.append(cosine_similarity(query_e,i))
    

    return txt_list[similarity.index(max(similarity))]
    
```
